```python
bins = [1960, 1970, 1980, 1990, 2000, 2010,2020]#6-intervals
labels=['1960-1970','1971-1980', '1981-1990', '1991-2000', '2001-2010','2011-2020']# 6-labels

df['year_range'] = pd.cut(df['year'], bins=bins, labels=labels)#new col based on bin and labels
#print(df)#has year-range in every data row
df_range_counts = df['year_range'].value_counts().sort_index()# counts total of each item in year-range.
print(df_range_counts)

plt.figure(figsize=(10, 6))
df_range_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Years in Ranges')
plt.xlabel('Year Range')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize = (12,10))
cert_counts = df['certification'].value_counts()
plt.pie(cert_counts, labels=cert_counts.index, autopct='%1.1f%%', startangle=90)
```
