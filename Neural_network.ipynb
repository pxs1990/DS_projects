# **Neural Network**
Deep learning is a subset of machine learning that involves the use of neural networks with multiple layers (deep neural networks) to automatically learn and make sense of data. The "deep" in deep learning refers to the depth of the neural networks, meaning they have several layers through which data is processed. It is inspired by the structure and functioning of the human brain.

Key aspects of deep learning include:

Neural Networks:

Deep learning is based on artificial neural networks, which are computational models inspired by the way biological neural networks in the human brain work.
Neural networks consist of interconnected nodes (artificial neurons) organized in layers, including an input layer, one or more hidden layers, and an output layer.

Multiple Layers:

A defining feature of deep learning is the use of deep neural networks with multiple hidden layers. Traditional machine learning models might have fewer layers or be shallower in comparison.
The depth of the network allows it to automatically learn hierarchical representations of data, capturing complex patterns and features.

Automated Feature Learning:

One of the strengths of deep learning is its ability to automatically learn relevant features from raw data. In traditional machine learning, feature engineering is often done manually, but deep learning algorithms can learn useful representations on their own.

Training Process:

Training a deep learning model involves feeding it with a large amount of labeled data and adjusting the weights of the connections between neurons during a process called backpropagation.
The learning process involves minimizing the difference between the model's predictions and the actual labels in the training data.

Applications:

Deep learning has shown remarkable success in various applications, including image and speech recognition, natural language processing, computer vision, and playing strategic games like Go.
Convolutional Neural Networks (CNNs) are commonly used in image-related tasks, Recurrent Neural Networks (RNNs) in sequential data, and Transformers in natural language processing.
Deep learning has become increasingly popular due to its ability to handle complex tasks and large datasets, leading to breakthroughs in various fields. However, it also requires substantial computational resources and data for effective training.

```python
import sklearn
from sklearn.datasets import load_breast_cancer
cancer_data=load_breast_cancer()#dict form
print(cancer_data)
```

```python
from sklearn.model_selection import train_test_split
X=cancer_data['data']
y=cancer_data['target']
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=.2)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
model = Sequential()
# Input layer
model.add(Dense(64, input_shape=(30,), activation='tanh'))
# hidden layer
model.add(Dense(32, activation='tanh'))# act fun introduces non-linearity
model.add(Dense(30, activation='relu'))
model.add(Dense(28, activation='relu'))
model.add(Dense(26, activation='relu'))
model.add(Dense(24, activation='relu'))
# out layer
# because this is a binary classification we use sigmoid function
model.add(Dense(1, activation='sigmoid'))
# Model compile
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X, y, epochs=20, batch_size=10)
loss_value, accuracy = model.evaluate(X, y)
print('Accuracy: %.2f' % (accuracy*100))
print("loss",loss_value)
```

**Optimizer**

In the context of deep learning, the optimizer is a crucial component of the training process for neural networks. The optimizer is responsible for updating the weights of the neural network based on the computed gradients during backpropagation, aiming to minimize the loss function. The loss function measures the difference between the predicted values and the actual values.

The choice of optimizer and its parameters can significantly impact the training process, convergence speed, and final performance of the model. Some commonly used optimizers include:

Stochastic Gradient Descent (SGD):

Traditional optimization algorithm.
Updates weights in the opposite direction of the gradient with a certain step size (learning rate).
Adam (Adaptive Moment Estimation):

An adaptive learning rate optimization algorithm.
Combines the advantages of both AdaGrad and RMSProp.
Dynamically adjusts the learning rates for each parameter based on past gradients.
RMSProp (Root Mean Square Propagation):

An adaptive learning rate optimization algorithm.
Divides the learning rate by an exponentially decaying average of squared gradients.
Adagrad (Adaptive Gradient Algorithm):

An adaptive learning rate optimization algorithm.
Adjusts the learning rates for each parameter based on the historical gradient information.
Adamax:

A variant of Adam that replaces the squared gradients in the denominator with the infinity norm of the gradients.
Nadam (Nesterov-accelerated Adaptive Moment Estimation):

An extension of Adam that incorporates Nesterov accelerated gradient.
Choosing the right optimizer depends on the specific characteristics of the problem at hand. Adam is a popular and often recommended choice for many tasks due to its adaptive learning rate properties. However, it's a good practice to experiment with different optimizers and learning rates to find the one that works best for a particular problem.

When using the compile method in Keras or TensorFlow, you specify the optimizer, loss function, and evaluation metrics. The optimizer's role is to minimize the specified loss function during training.

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
# Input layer
model.add(Dense(64, input_shape=(30,), activation='relu'))
# hidden layer
model.add(Dense(32, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(28, activation='relu'))
model.add(Dense(26, activation='relu'))
model.add(Dense(24, activation='relu'))
# out layer
# because this is a binary classification we use sigmoid function
model.add(Dense(1, activation='sigmoid'))
# Model compile
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=15, batch_size=20)
#model.fit(X, y, epochs=20, batch_size=10)
loss_test, accuracy_test = model.evaluate(X_test, y_test)
loss_train, accuracy_train = model.evaluate(X_train, y_train)
print('Accuracy on test set: %.2f' % (accuracy_test*100))
print('Accuracy on train set: %.2f' % (accuracy_train*100))
print("loss test ",loss_test)
print("loss train ",loss_train)
```

```python
from sklearn.datasets import load_iris
import pandas as pd
iris_data=load_iris()
iris_data
```

```python
iris_data.keys()
```

```python
from keras.utils import to_categorical
X=iris_data['data']
y=iris_data['target']
y=to_categorical(y,num_classes=3)
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=.2, random_state=0)
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
model = Sequential()
# Input layer
model.add(Dense(64, input_shape=(4,), activation='relu'))
# hidden layer
model.add(Dense(32, activation='relu'))
model.add(Dense(30, activation='relu'))
model.add(Dense(28, activation='relu'))
model.add(Dense(26, activation='relu'))
model.add(Dense(24, activation='relu'))
# out layer
# because this is a multiclass classification we use softmax function
model.add(Dense(3, activation='softmax'))
# Model compile
optim=Adam(learning_rate=.001)
model.compile(loss='categorical_crossentropy', optimizer=optim, metrics=['accuracy'])
model.fit(X_train, y_train, epochs=15, batch_size=20)
#model.fit(X, y, epochs=20, batch_size=10)
loss_test, accuracy_test = model.evaluate(X_test, y_test)
loss_train, accuracy_train = model.evaluate(X_train, y_train)
print('Accuracy on test set: %.2f' % (accuracy_test*100))
print('Accuracy on train set: %.2f' % (accuracy_train*100))
print("loss test ",loss_test)
print("loss train ",loss_train)
```
